### Import Libraries

```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("movies.csv")
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
df_clean = df.dropna()
print(f"\nFilas después de limpiar: {len(df_clean)}")
```

```python
df_clean.isnull().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
df.hist(figsize=(10, 10))
plt.show()
```

```python
highest_score_movie = df.loc[df["score"].idxmax()]
print(highest_score_movie[["name", "score", "year", "director"]])
```

```python

max_score = df_clean["score"].max()
best_movie = df_clean[df_clean["score"] == max_score]

print("\nPelícula con mejor puntuación:")
print(best_movie[["name", "year", "score", "genre", "director"]].to_string(index=False))
```

```python

corr_matrix = df_clean[["score", "votes", "budget", "gross", "runtime"]].corr()

print("\nMatriz de correlación:")
print(corr_matrix)

import seaborn as sns
plt.figure(figsize=(10,10))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", center=0)
plt.title("Correlación entre Variables")
plt.show()
```

```python

genre_counts = df_clean["genre"].value_counts()

print("\nPelículas por género:")
print(genre_counts)

plt.figure(figsize=(10, 10))
genre_counts.plot(kind="bar", color="pink")
plt.title("Distribución de Películas por Género")
plt.xlabel("Género")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(10, 10))
plt.scatter(df_clean["budget"], df_clean["gross"], alpha=0.5, color="pink")
plt.title("Relación entre Presupuesto y Recaudación")
plt.xlabel("Presupuesto (millones $)")
plt.ylabel("Recaudación (millones $)")
plt.grid(True)
plt.show()


df_clean["ROI"] = (df_clean["gross"] - df_clean["budget"]) / df_clean["budget"]
print("\nTop 5 películas con mejor ROI:")
print(df_clean.sort_values("ROI", ascending=False)[["name", "ROI"]].head().to_string(index=False))
```

```python

yearly_score = df_clean.groupby("year")["score"].mean()

plt.figure(figsize=(14, 6))
yearly_score.plot()
plt.title("Evolución de la Puntuación Promedio por Año")
plt.xlabel("Año")
plt.ylabel("Puntuación Promedio")
plt.grid(True)
plt.show()
```

```python

df_clean.to_csv("movies_clean.csv", index=False)
print("\nDataset limpio guardado como 'movies_clean.csv'")
```
